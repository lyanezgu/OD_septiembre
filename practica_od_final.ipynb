Fecha de descarga de datos: 1/08/22

```python
datos_aire = pd.read_csv('calidad_aire_datos_meteo_mes.csv', sep=';')
datos_aire.head(25)
```

Cambiamos los tipos de algunas variables y en algunas de ellas añadimos ceros en el caso en que las varibales no tengan 3 dígitos.

```python
datos_aire['provincia'] = pd.Series(datos_aire['provincia']).astype(str)
datos_aire['municipio'] = pd.Series(datos_aire['municipio']).astype(str).str.zfill(3)
datos_aire['estacion'] = pd.Series(datos_aire['estacion']).astype(str).str.zfill(3)
datos_aire['estacion_codigo'] = datos_aire['provincia'] + datos_aire['municipio'] + datos_aire['estacion']
datos_aire['magnitud'] = pd.Series(datos_aire['magnitud']).astype(str)
datos_aire.head()
```

Movemos la variable estacion_codigo a la primera posición 

```python
estacion_codigo= datos_aire['estacion_codigo']
#estacion_codigo.head()
datos_aire=datos_aire.drop(['estacion_codigo'], axis=1)
datos_aire.insert(0,'estacion_codigo',estacion_codigo)
datos_aire = datos_aire.sort_values(by=['estacion_codigo'], ascending=True)
datos_aire.head()
```

Cargamos los datos del CSV calidad_aire_estaciones y seleccionamos las tres variables que se nos indica en la práctica (datos de nombre de la estación, latitud y longitud (UTM))

```python
nombres_colum = ['estacion_codigo', 'estacion_coord_UTM_ETRS89_x', 'estacion_coord_UTM_ETRS89_y']
datos_estaciones = pd.read_csv('calidad_aire_estaciones.csv', sep=';', encoding='latin-1', usecols=nombres_colum)
datos_estaciones['estacion_codigo'] = pd.Series(datos_estaciones['estacion_codigo']).astype(str)
datos_estaciones.head()
```

Juntamos los ficheros csv uniendolos a través de la columna "estacion_codigo".

Creamos la variable date_observation la cual esta formada por punto_muestreo, año, mes y día. Movemos la variable date_observation a la primera posición.

```python
df_aire_esta= pd.merge(datos_aire, datos_estaciones, how="left", on="estacion_codigo")
df_aire_esta['date_observation'] = pd.Series(df_aire_esta['punto_muestreo']).astype(str)+pd.Series(df_aire_esta['ano']).astype(str) + pd.Series(df_aire_esta['mes']).astype(str).str.zfill(2) + pd.Series(df_aire_esta['dia']).astype(str).str.zfill(2)
date_observation= df_aire_esta['date_observation']
df_aire_esta=df_aire_esta.drop(['date_observation'], axis=1)
df_aire_esta.insert(0,'date_observation',date_observation)
df_aire_esta = df_aire_esta.sort_values(by=['estacion_codigo'], ascending=True)
df_aire_esta.head()
```

Generamos el grafo RDF definiendo los datos semánticos anotados en tripletas.

```python
for index, row in df_aire_esta.iterrows():
    estacion_codigo= URIRef(row['estacion_codigo'])
    date_observation = URIRef(row['date_observation'])
    punto_muestreo= URIRef(row['punto_muestreo'])
    g.add((estacion_codigo, has_provincia, Literal(row['provincia'])))
    if row['municipio'] in dic_municipios.keys():
        key= row['municipio']
        val= dic_municipios[key]
        uri_municipio=URIRef(val)
        g.add((estacion_codigo, has_municipio,uri_municipio))

    g.add((estacion_codigo, has_estacion, Literal(row['estacion'])))
    g.add((estacion_codigo, has_latitude, Literal(row['estacion_coord_UTM_ETRS89_x'])))
    g.add((estacion_codigo, has_longitude, Literal(row['estacion_coord_UTM_ETRS89_y'])))
    g.add((estacion_codigo, has_punto_muestreo, punto_muestreo))
    g.add((punto_muestreo, has_magnitud, Literal(row['magnitud'])))
    g.add((punto_muestreo, has_date_observation, date_observation)) # date (punto_muestreo + fecha)
    g.add((date_observation, has_h01, Literal(row['h01']))) 
    g.add((date_observation, has_v01, Literal(row['v01'])))
    g.add((date_observation, has_h01, Literal(row['h02']))) 
    g.add((date_observation, has_v01, Literal(row['v02'])))
    g.add((date_observation, has_h01, Literal(row['h03']))) 
    g.add((date_observation, has_v01, Literal(row['v03'])))
    g.add((date_observation, has_h01, Literal(row['h04']))) 
    g.add((date_observation, has_v01, Literal(row['v04'])))
    g.add((date_observation, has_h01, Literal(row['h05']))) 
    g.add((date_observation, has_v01, Literal(row['v05'])))
    g.add((date_observation, has_h01, Literal(row['h06']))) 
    g.add((date_observation, has_v01, Literal(row['v06'])))
    g.add((date_observation, has_h01, Literal(row['h07']))) 
    g.add((date_observation, has_v01, Literal(row['v07'])))
    g.add((date_observation, has_h01, Literal(row['h08']))) 
    g.add((date_observation, has_v01, Literal(row['v08'])))
    g.add((date_observation, has_h01, Literal(row['h09']))) 
    g.add((date_observation, has_v01, Literal(row['v09'])))
    g.add((date_observation, has_h01, Literal(row['h10']))) 
    g.add((date_observation, has_v01, Literal(row['v10'])))
    g.add((date_observation, has_h01, Literal(row['h11']))) 
    g.add((date_observation, has_v01, Literal(row['v11'])))
    g.add((date_observation, has_h01, Literal(row['h12']))) 
    g.add((date_observation, has_v01, Literal(row['v12'])))
    g.add((date_observation, has_h01, Literal(row['h13']))) 
    g.add((date_observation, has_v01, Literal(row['v13'])))
    g.add((date_observation, has_h01, Literal(row['h14']))) 
    g.add((date_observation, has_v01, Literal(row['v14'])))
    g.add((date_observation, has_h01, Literal(row['h15']))) 
    g.add((date_observation, has_v01, Literal(row['v15'])))
    g.add((date_observation, has_h01, Literal(row['h16']))) 
    g.add((date_observation, has_v01, Literal(row['v16'])))
    g.add((date_observation, has_h01, Literal(row['h17']))) 
    g.add((date_observation, has_v01, Literal(row['v17'])))
    g.add((date_observation, has_h01, Literal(row['h18']))) 
    g.add((date_observation, has_v01, Literal(row['v18'])))
    g.add((date_observation, has_h01, Literal(row['h19']))) 
    g.add((date_observation, has_v01, Literal(row['v19'])))
    g.add((date_observation, has_h01, Literal(row['h20']))) 
    g.add((date_observation, has_v01, Literal(row['v20'])))
    g.add((date_observation, has_h01, Literal(row['h21']))) 
    g.add((date_observation, has_v01, Literal(row['v21'])))
    g.add((date_observation, has_h01, Literal(row['h22']))) 
    g.add((date_observation, has_v01, Literal(row['v22'])))
    g.add((date_observation, has_h01, Literal(row['h23']))) 
    g.add((date_observation, has_v01, Literal(row['v23'])))
    g.add((date_observation, has_h01, Literal(row['h24']))) 
    g.add((date_observation, has_v01, Literal(row['v24'])))
    g.add((date_observation, has_ano, Literal(row['ano'])))
    g.add((date_observation, has_mes, Literal(row['mes'])))
    g.add((date_observation, has_dia, Literal(row['dia'])))
    
    
g.serialize(destination ="datossemanticos.rdf", format = "xml")
print("generado") 
    
```
